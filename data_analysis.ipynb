```python
import pandas as pd


df = pd.read_csv("/content/train_split_00.csv")

# Display the dataset, its first few rows, basic information, and summary statistics
print(df)
print(df.head())
print(df.info())
print(df.describe())


# Check if the 'minute' column exists and then perform calculations
if 'minute' in df.columns:
    print("Column 'minute' found in Goal Scores dataset")
    mean_value = df ['minute'].mean()
    median_value = df ['minute'].median()
    mode_value = df['minute'].mode().iloc[0] if not df['minute'].mode().empty else None
    variance_value = df['minute'].var()
    std_value = df['minute'].std()
    range_value = df['minute'].max() - df['minute'].min()

    print("Goal Scores - Minute Statistics:")
    print(f"Mean: {mean_value}")
    print(f"Median: {median_value}")
    print(f"Mode: {mode_value}")
    print(f"Variance: {variance_value}")
    print(f"Standard Deviation: {std_value}")
    print(f"Range: {range_value}")

else:
    print("Column 'minute' not found in Goal Scores dataset")
```
